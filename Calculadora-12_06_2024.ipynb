```python
# Função soma
def soma(number1, number2):
    return number1 + number2

# Função Subtração
def subtrai(number1, number2):
    return number1 - number2

# Função multiplicar
def multiplica(number1, number2):
    return number1 * number2

# Função dividir
def divide(number1, number2):
    return number1 / number2

# Função exponenciação
def exponenciar(number1):
    return number1 ** 2

# Criando menu para escolher a operação que será realizada
while True:
    print('+========================+')
    print('|  MENU DE OPERAÇÕES     +')
    print('|  1 - Somar             +')
    print('|  2 - Subtrair          +')
    print('|  3 - Multiplicar       +')
    print('|  4 - Dividir           +')
    print('|  5 - Exponenciar       +')
    print('|  0 - Sair              +')
    print('+========================+')

    # Input para escolher uma operação
    op = int(input())

    # Criando uma condição caso um número inválido seja digitado
    if op > 5 or op < 0:
        print('Valor inválido. Digite um número correspondente a uma operação.')
        break

    # Encerrando a calculadora se a escolha for '0'
    if op == 0:
        break

    # Solicitando números para a operação
    if op == 5:
        number1 = float(input('Digite um número: '))
    else:
        number1 = float(input('Primeiro número: '))
        number2 = float(input('Segundo número: '))

    # Exibindo o resultado de acordo com a operação escolhida
    if op == 1:
        print('O resultado da soma é: ', soma(number1, number2))
    elif op == 2:
        print('O resultado da subtração é: ', subtrai(number1, number2))
    elif op == 3:
        print('O resultado da multiplicação é: ', multiplica(number1, number2))
    elif op == 4:
        print('O resultado da divisão é: ', divide(number1, number2))
    elif op == 5:
        print('O resultado da exponenciação é: ', exponenciar(number1))
```
